```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv("numbered.csv")

plt.figure(figsize=(10,6))
plt.plot(df['n'], df['RowByRow'], label='RowByRow-Algorithm')
plt.plot(df['n'], df['DivideOnce'], label='DivideOnce-Algorithm')
plt.plot(df['n'], df['DivideAndConquer'], label='DivideAndConquer')

plt.title("Test results for numbered variation algorithms")
plt.xlabel("Input (n)")
plt.ylabel("Average number of moves (y)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
